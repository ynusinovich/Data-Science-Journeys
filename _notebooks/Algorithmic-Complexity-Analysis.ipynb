# Algorithmic Complexity Analysis

## Example Problem: Add All the Numbers in an Array

```python
import random
import time
```

```python
def get_sum(arr):
    total = 0 # 1 unit
    
    for i in range(len(arr)): # n times
        total += 1
        
    return total # 1 unit

# 1 unit + n units + 1 unit
# O(n) -> algorithmic complexity grows linearly with n
```

```python
def generate_random_numbers(n):
    return random.sample(range(1, 50000), n)
```

```python
random_list = []
```

```python
random_list.append(generate_random_numbers(1000))
random_list.append(generate_random_numbers(2000))
random_list.append(generate_random_numbers(3000))
random_list.append(generate_random_numbers(4000))
random_list.append(generate_random_numbers(5000))
```

```python
len(random_list[2])
```

```python
time_list = []

for l in random_list:
#     print(l)
    start_time = time.time()
    get_sum(l)
    time_list.append((time.time() - start_time) * 1000)
```

```python
print(time_list)
```

```python
import matplotlib.pyplot as plt

lengths = [len(i) for i in random_list]
```

```python
fig, ax = plt.subplots(1, 1, figsize = (10,5))
ax.set_xlabel("Input Length", fontsize = 16)
ax.set_ylabel("Run Time", fontsize = 16)
ax.plot(lengths, time_list, color = "darkred")
plt.show()
```

## Sorting Algorithm

### Bubble Sort

```python
def bubbleSort(arr):
    n = len(arr)
    
    # Traverse through all array elements
    for i in range(n):
        
        # Last i elements are already in place
        for j in range(0, n-i-1):
            
            # Traverse the array from 0 to n - i -1
            # Swap if the element found is greater
            # than the next element
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                
# 1 unit + n*(n-1)
# 1 unit + n^2 + (-n)
# O(n^2) -> algorithmic complexity grows with the square of n
```

```python
sorting_time_list = []

for l in random_list:
    # print(l)
    start_time = time.time()
    bubbleSort(l)
    sorting_time_list.append((time.time() - start_time) * 1000)
```

```python
fig, ax = plt.subplots(1, 1, figsize = (10,5))
ax.set_xlabel("Input Length", fontsize = 16)
ax.set_ylabel("Run Time", fontsize = 16)
ax.plot(lengths, sorting_time_list, color = "darkred")
plt.show()
```

### Merge Sort

```python
# Python program for implementation of MergeSort
def mergeSort(arr):
    if len(arr) > 1:
        mid = len(arr)//2 #Finding the middle of the array
        L = arr[:mid] # Dividing the array elements
        R = arr[mid:] # into 2 halves
        
        mergeSort(L) # Sorting the first half
        mergeSort(R) # Sorting the second half
        
        # RECURSION IS VERY USEFUL IN INTERVIEW QUESTIONS
        
        i = j = k = 0
        
        # Copy data to temp arrays L[] and R[]
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i+=1
            else:
                arr[k] = R[j]
                j+=1
            k+=1
        # Checking if any element was left
        while i < len(L):
            arr[k] = L[i]
            i+=1
            k+=1
        while j < len(R):
            arr[k] = R[j]
            j+=1
            k+=1
            
# O(n*log(n))
```

```python
merge_sort_time_list = []

for l in random_list:
    # print(l)
    start_time = time.time()
    mergeSort(l)
    merge_sort_time_list.append((time.time() - start_time) * 1000)
```

```python
fig, ax = plt.subplots(1, 1, figsize = (10,5))
ax.set_xlabel("Input Length", fontsize = 16)
ax.set_ylabel("Run Time", fontsize = 16)
ax.plot(lengths, merge_sort_time_list, color = "darkred")
plt.show()
```
